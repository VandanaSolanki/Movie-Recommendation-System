**Extract** name of genre

```python
import ast
```

```python
def convert(x):
  l=[]
  for i in ast.literal_eval(x):
    l.append(i['name'])
  return l
```

```python
movies['genres']=movies['genres'].apply(convert)
movies['genres']
```

**Extract** name of keywords

```python
movies['keywords']=movies['keywords'].apply(convert)
movies['keywords']
```

**Extract** 5 names from cast

```python
def convert3(x):
  l=[]
  c=0
  for i in ast.literal_eval(x):
    if c!=5:
      l.append(i['name'])
      c+=1
    else:
      break
  return l
```

```python
movies['cast']=movies['cast'].apply(convert3)
movies['cast']
```

```python
movies.head()
```

**Extract** job director from crew

```python
def get_director(x):
  l=[]
  for i in ast.literal_eval(x):
    if i['job']=='Director':
      l.append(i['name'])
      break
  return l
```

```python
movies['crew']=movies['crew'].apply(get_director)
movies['crew']
```

**Modify** overview

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies['overview']
```

```python
movies.head()
```

**Remove Space**

```python
from os import replace
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies.head()
```

**Tags = genres + keywords + overview + cast + crew**

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
movies.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
new_df.head()
```

**steming similar words**

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  l1=[]
  for i in text.split():
    l1.append(ps.stem(i))
  return " ".join(l1)
```

```python
# Example of steming
print(ps.stem('dance'))
print(ps.stem('dancing'))
print(ps.stem('danced'))
```

```python
new_df['tags']=new_df['tags'].apply(stem)
new_df['tags']
```

**converting words into vector**

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
cv.get_feature_names()
```

**Similarity of one movie to others (closeness in the vectors)**

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

Index of the movie on the basis of title

```python
movie_index=new_df[new_df['title']=='Batman Begins'].index[0]
movie_index
```

```python
#similarity with index
list(enumerate(similarity[0]))
```

```python
#Sort on the basis of highest similarity (top 5)
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index=new_df[new_df['title']==movie].index[0]
  distances=similarity[movie_index]
  movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movie_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
